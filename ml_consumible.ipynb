```python
import pandas as pd 
from pycaret.regression import *
```

```python
# Se carga el arhivo parquet
df= pd.read_parquet('dataset\\verde_final.parquet')
```

```python
# Se configura la inicializaci√≥n de Pycaret
s = setup(df,target='total_amount',session_id=123)
```

```python
# Se crea el modelo Logistic Regression 
lr = create_model('lr')
```

```python
# Se visualizan la importancia de las variables seleccionadas para el modelo
plot_model(lr,plot='feature')
```

```python
# Se visualiza el margen de error para este modelo
plot_model(lr,plot='error')
```

```python
# Se analizan los residuales en un gr√°fico
plot_model(lr,plot='residuals')

```

```python
# Se guarda el modelo para hacer el deploy en streamlit
save_model(lr,model_name='deployment1')

```
